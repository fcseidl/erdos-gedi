```python
import pandas as pd
import os
```

```python
granules = os.listdir("data")
print(len(granules))
df = pd.concat([pd.read_csv("data/" + g) for g in granules if g.endswith(".csv")])
df
```

```python
count_empty = 0
for g in granules:
    if g.endswith(".csv"):
        path = os.path.join("data", g)
        granuledf = pd.read_csv(path)
        if granuledf.shape[0] == 0:
            count_empty += 1
            os.remove(path)
print(f"Removed {count_empty} empty .csv files.")
```
